### 2.1 Изучите распределение домов от наличия вида на набережную

```python

data = df['waterfront'].value_counts()
data.index = ['no', 'yes']
data.head()
```

```python

plt.figure(figsize=(6, 4))
plt.pie(data, autopct='%.2f%%')
plt.title('Доля домов с наличием вида на набережную')
plt.legend(data.index);
```

Вывод: Доля домов с видом на набережную составляет всего 0,75%, что может быть обусловлено высокой стоимостью таких домов.

```python
import numpy as np
```

```python
corr_matrix = df.corr()
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
```

```python
plt.figure(figsize=(14, 12))
sns.set(font_scale=1.4)
sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm')
plt.title('Матрица корреляций');
```

Вывод: по матрице корреляций можно утверждать, что наибольшая взаимосвязь у цен домов отслеживается с размерностью жилой площади и с оценкой конструкции здания и дизайна. А с наличием вида на набережную и насколько хороший вид взаимосвязь всего 0,3 и 0,4 соответственно. Взаимосвязь общей площади и года реновации со стоимостью также подтвердилась: в обоих случаях нулевая.
